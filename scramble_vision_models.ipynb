```python
import glob
from scipy import misc
from psychopy_ext import models
import matplotlib.pyplot as plt
%matplotlib inline

def simple_implot(im, crop=(110, 440, 70, 360), figsize=(6,6)):
    """
    simple wrapper for matplotlib.pyplot.imshow(): crops the image to 'zoom in' on the face,
    creates a figure of the given figsize, turns of axes and, if using a Qt-Backend, moves the
    plot to the second screen.
    """
    if crop:
        y,h,x,w = crop
        im = im[y:y+h, x:x+w]
    plt.figure(figsize=figsize)
    plt.axis('off')
    plt.imshow(im)
    plt.show()
```
